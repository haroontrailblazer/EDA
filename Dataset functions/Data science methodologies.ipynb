```python
import pandas as pd
series = pd.Series([1, 1, 1, 1, 3, 3, 3, 6, 7, 7, 7, 54, 54, 100, 100])
print('Frequency:\n', series.value_counts())
```

```python
import scipy.stats as stats
print('Mean:',np.mean(series))
print('Median:',np.median(series))
print('Mode:',stats.mode(series))
```

```python
print('numpy:')
print('Variance:',np.var(series))
print('Standard Deviation:',np.std(series))
print('Range:',np.ptp(series))
print('pandas:')
print(pd.Series(series).describe())
```

```python
import seaborn as sns

mu = 0
sigma = 1
data = np.random.normal(mu, sigma, 1000)


sns.histplot(data, bins=30, kde=True, color='g', stat='density', alpha=0.6)

plt.xlabel('Data values')
plt.ylabel('Probability density')
plt.title('Normal Distribution Curve')
plt.show()
```

```python
x=np.random.rand(100)
y=2*x+np.random.normal(0,0.1,100)
plt.scatter(x,y,alpha=0.5,color='red')
plt.show()
```

```python
print('Numpy-Correlation:', np.corrcoef(x, y))
print('Pandas-Correlation:', pd.DataFrame({'x': x, 'y': y}).corr())
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = np.random.rand(1000, 1)
y = 2 * x + np.random.normal(0, 0.1, (1000, 1))
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

plt.scatter(x_test, y_test, color='blue', alpha=0.5)
plt.plot(x_test, y_pred, color='red', linewidth=2)
plt.show()
```
