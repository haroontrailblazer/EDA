```python
import matplotlib.pyplot as plt 
import numpy as np 
# Spiral controls the number of Spiral 
spiral = 5 
# data 
zline = np.linspace(0,5,100) 
xline = np.cos(spiral * zline) 
yline = np.sin(spiral * zline) 
ax = plt.axes(projection ='3d') 
ax.plot3D(xline, yline, zline, lw=5); 
ax.set_xlabel('X - Axis') 
ax.set_ylabel('Y - Axis') 
ax.set_zlabel('Z - Axis') 
plt.show()
```

```python
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
import numpy as np 
# Generate sample data 
np.random.seed(0) 
x = np.random.rand(100) 
y = np.random.rand(100) 
z = np.random.rand(100) 
# Create a figure and a 3D axes 
fig = plt.figure(figsize=(10, 7)) 
ax = fig.add_subplot(111, projection='3d') 
# Create a 3D scatter plot 
scatter = ax.scatter(x, y, z, c='blue', marker='o') 
# Set labels and title 
ax.set_xlabel('X-axis') 
ax.set_ylabel('Y-axis') 
ax.set_zlabel('Z-axis') 
ax.set_title('3D Scatter Plot') 
# Show the plot 
plt.show() 
```

4.Surface plot

```python
import numpy as np
import matplotlib.pyplot as mp
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
x1,y1=np.meshgrid(x,y)
z=x1**2+y1**2
a=mp.subplot(projection='3d')
a.plot_surface(x1,y1,z,color='b')
mp.show()
```

4.Surface plot using color map

```python
import numpy as np
import matplotlib.pyplot as mp
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
x1,y1=np.meshgrid(x,y)
z=x1**2+y1**2
cl=mp.colormaps()
for i in range(0,len(cl)):
    print(i+1 ,cl[i])
    a=mp.subplot(projection='3d')
    a.plot_surface(x1,y1,z,cmap=cl[i])
    mp.show()
```
