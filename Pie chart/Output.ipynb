1.Basic pie chart

```python
import matplotlib.pyplot as mp

# Data
marks = [10, 20, 30, 40, 50, 100]
subjects = ['Maths', 'Physics', 'Chemistry', 'English', 'Computer', 'Avg']
colors = ['red', 'green', 'blue', 'purple', 'pink', 'orange']
explode = [0.1, 0.3, 0.01, 0.2, 0.1, 0.1]

# Set the style
mp.style.use('dark_background')

# Create a figure with a specific size and resolution
mp.figure(figsize=(12, 12), dpi=100)

# Subplot 1: Basic Pie Chart
ax1 = mp.subplot(2, 2, 1)
ax1.pie(marks, labels=subjects)
ax1.set_title('1.Basic Pie Chart')

# Subplot 2: Pie Chart with Percentages
ax2 = mp.subplot(2, 2, 2)
ax2.pie(marks, labels=subjects, autopct='%0.1f%%')
ax2.set_title('2.pie Chart with Percentages')

# Subplot 3: Pie Chart with Colors
ax3 = mp.subplot(2, 2, 3)
ax3.pie(marks, labels=subjects, autopct='%0.1f%%', colors=colors)
ax3.set_title('3.Pie Chart with Colors')

# Subplot 4: Pie Chart with Explode
ax4 = mp.subplot(2, 2, 4)
ax4.pie(marks, labels=subjects, autopct='%0.1f%%', colors=colors, explode=explode)
ax4.set_title('4.Pie Chart with Explode')

# Adjust layout to prevent overlap
mp.tight_layout()

# Show the plot
mp.show()
```

2.Pie chart with percentage

```python
import matplotlib.pyplot as mp
mp.style.use('dark_background')
subjects = ['Maths', 'Physics', 'Chemistry', 'English', 'Computer','Avg']
marks = [10, 20, 30, 40, 50,100]
mp.pie(marks,labels=subjects,autopct='%0.1f%%')
mp.title('Marks of Students')
mp.show()
```

3.Pie chart with color customization

```python
import matplotlib.pyplot as mp
mp.style.use('dark_background')
subjects = ['Maths', 'Physics', 'Chemistry', 'English', 'Computer','Avg']
marks = [10, 20, 30, 40, 50,100]
cl=['red','green','blue','purple','pink','orange']
mp.pie(marks,labels=subjects,autopct='%0.1f%%',colors=cl)
mp.title('Marks of Students')
mp.show()
```
