1.Basic bubble chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)
sizes = np.random.rand(10) * 1000
colors = np.random.rand(10)

# Basic Bubble Chart
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title('Basic Bubble Chart')
```

2.Bubble chart with color

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)
sizes = np.random.rand(10) * 1000
colors = np.random.rand(10)
# Bubble Chart with Color
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title('Bubble Chart with Color')
```

3.Bubble chart with edge color

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)
sizes = np.random.rand(10) * 1000
colors = np.random.rand(10)
#Bubble Chart with Edge Colors
plt.scatter(x, y, s=sizes, alpha=0.5, edgecolors='k', linewidth=2)
plt.title('Bubble Chart with Edge Colors')
```

4.bubble chart with different markers

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)
sizes = np.random.rand(10) * 1000
colors = np.random.rand(10)
# Bubble Chart with Different Markers
markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h']
for i in range(len(x)):
    plt.scatter(x[i], y[i], s=sizes[i], alpha=0.5, marker=markers[i % len(markers)], label=f'Point {i+1}')
plt.legend()
plt.title('Bubble Chart with Different Markers')
```
