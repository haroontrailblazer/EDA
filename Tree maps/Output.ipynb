1.Basic tree map

```python
import matplotlib.pyplot as plt
import squarify

# Data
sizes = [50, 25, 12, 6, 4, 3]
labels = ['A', 'B', 'C', 'D', 'E', 'F']

# Plot
plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=labels, alpha=0.7)
plt.axis('off')
plt.show()
```

2.Tree map with specifiied colours

```python
import matplotlib.pyplot as plt
import squarify

# Data
sizes = [50, 25, 12, 6, 4, 3]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
colors = ['red', 'green', 'blue', 'purple', 'orange', 'pink']

# Plot
plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=0.7)
plt.axis('off')
plt.show()
```

3.labeling values

```python
import matplotlib.pyplot as plt
import squarify

# Data
sizes = [50, 25, 12, 6, 4, 3]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
value_labels = [f'{label}\n{size}' for label, size in zip(labels, sizes)]

# Plot
plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=value_labels, alpha=0.7)
plt.axis('off')
plt.show()
```

4.Nested tree valuues

```python
import matplotlib.pyplot as plt
import squarify

# Data
sizes = [50, 25, 12, 6, 4, 3]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
nested_sizes = [30, 20, 10, 5, 3, 2]
nested_labels = ['A1', 'A2', 'B1', 'B2', 'C1', 'C2']

# Plot
plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=labels, alpha=0.7)
squarify.plot(sizes=nested_sizes, label=nested_labels, alpha=0.5, color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
plt.axis('off')
plt.show()
```
